3. Explain the following terms

a) The difference in np.array, np.asarray and np.asanyarray.

b) The difference between Deep copy and shallow copy.

 np.array, creates a new NumPy array from a given object (like lists, tuples, etc.), it always return a new array and copies the data from the original array.

 np.asarray, Converts input to an array but does not copy the data if the input is an existing array, it returns the same array (no new array is created).

 np.asanyarray is similar to np.asarray, but it allows for input of any array type (including subclasses of ndarray, like MaskedArray).

Shallow copy:- Creates a new object, but does not create copies of nested objects within it. It only copies references to the nested objects. So if you modify something in a modified array, the change will reflect in the original array as well.

Deep copy:- Creates a new object and copies everything within it, it doesn't create any references, so making any changes in the new array won't change anything in the original array.

Question 17

```python
import seaborn as sns

x = np.random.randn(100)
y = np.random.randn(100)

colors = np.where((x > 0) & (y > 0), 'blue', 
          np.where((x < 0) & (y > 0), 'green', 
          np.where((x < 0) & (y < 0), 'red', 'orange')))

plt.figure(figsize=(8,6))
scatter = sns.scatterplot(x=x, y=y, hue=colors, palette=['blue', 'green', 'red', 'orange'], legend='full')

handles, labels = scatter.get_legend_handles_labels()
scatter.legend(handles, ['Q1 (blue)', 'Q2 (green)', 'Q3 (red)', 'Q4 (orange)'], title="Quadrants")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')

plt.show()
```

```python

output_notebook()


categories = ['A', 'B', 'C', 'D', 'E']
values = [np.random.randint(10, 100) for _ in categories]
source = ColumnDataSource(data=dict(categories=categories, values=values))
p = figure(x_range=categories, title="Random Categorical Bar Chart", 
           x_axis_label='Categories', y_axis_label='Values')

p.vbar(x='categories', top='values', width=0.7, source=source)

hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)
show(p)
```
